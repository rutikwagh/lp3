```python
x=2
lr=0.01
precision=0.000001
previos=1
max_iter=10000
iters=0
gf= lambda x: (x + 3)**2
```

```python
gd=[]
```

```python
while precision<previos and iters<max_iter:
    prev=x
    x=x-lr*gf(prev)
    previos=abs(x-prev)
    iters+=1
    print("iterations : ",iters,"  values: ",x)
    gd.append(x)
    
```

```python
print("local minima : ",x)
```

```python
import matplotlib.pyplot as plt
plt.plot(gd)
```
